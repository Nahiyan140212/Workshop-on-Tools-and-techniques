# Introduction to Matplotlib
Matplotlib is the fundamental plotting library in Python. It provides low-level control over plot elements, a variety of plot types (line, bar, scatter, histogram, etc.).

```python
import matplotlib.pyplot as plt

# Create sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 8, 7]

# Create a line plot
plt.figure(figsize=(6,4))
plt.plot(x, y, marker='o', linestyle='-', color='blue')
plt.title('Basic Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].plot(x, y, marker='o', color='red')
axes[0].set_title('Subplot 1')

axes[1].bar(x, y, color='green')
axes[1].set_title('Subplot 2')

plt.tight_layout()
plt.show()
```

# Histogram and Density Plots
Histograms are useful for understanding the distribution of a continuous variable. You can also overlay a KDE (Kernel Density Estimate) curve for a smoother distribution view.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
np.random.seed(0)
data = np.random.normal(loc=50, scale=10, size=500)

plt.figure(figsize=(6, 4))
plt.hist(data, bins=20, alpha=0.7, color='blue', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# For a density plot using Matplotlib
plt.figure(figsize=(6, 4))
count, bins, ignored = plt.hist(data, bins=20, density=True, alpha=0.2, color='gray', edgecolor='black')
plt.plot(bins, 1/(10 * np.sqrt(2 * np.pi)) * np.exp( - (bins - 50)**2 / (2 * 10**2) ), 
         linewidth=2, color='red')
plt.title('Histogram with Theoretical Density')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.show()
```

# Scatter plot

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)  # for color mapping

plt.figure(figsize=(6, 4))
plt.scatter(x, y, c=colors, cmap='viridis', s=100, alpha=0.7, edgecolors='black')
plt.colorbar(label='Color scale')
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

# Plot from a dataset

```python
import pandas as pd

data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Sales_A': [100, 120, 90, 130, 150],
    'Sales_B': [80, 85, 95, 110, 120],
    'Sales_C': [50, 60, 40, 60, 65]
}

df = pd.DataFrame(data)
print(df)
```

# Line plot

```python
df.plot(x='Month', y=['Sales_A', 'Sales_B', 'Sales_C'], kind='line', 
        marker='o', figsize=(8, 4), title='Sales Over Months')
plt.ylabel('Sales')
plt.show()
```

# Bar Plot

```python
df.plot(x='Month', y=['Sales_A', 'Sales_B', 'Sales_C'], kind='bar', figsize=(8, 4))
plt.title('Sales Over Months - Bar Chart')
plt.ylabel('Sales')
plt.show()
```

# Box Plot

```python
df[['Sales_A', 'Sales_B', 'Sales_C']].plot(kind='box', figsize=(6, 4))
plt.title('Box Plot of Sales Data')
plt.ylabel('Sales')
plt.show()
```

# Scatter Plot

```python
# We'll create a new DataFrame for this example
np.random.seed(0)
df_scatter = pd.DataFrame({
    'Height': np.random.normal(170, 10, 50),
    'Weight': np.random.normal(70, 15, 50)
})

df_scatter.plot(kind='scatter', x='Height', y='Weight', color='green', edgecolor='black')
plt.title('Height vs Weight')
plt.show()
```
